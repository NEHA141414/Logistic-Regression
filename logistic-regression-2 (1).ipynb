```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_train=pd.read_csv("/kaggle/input/logistic-regression/titanic/train.csv")
```

```python
x_train.head()
```

```python
x_train.columns
```

```python
l1=['PassengerId','Name','Ticket']
```

```python
x_train.drop(l1,axis=1,inplace=True)
```

```python
x_train.head()
```

```python
x_train.info()
```

```python
l2=['Cabin']
x_train.drop(l2,axis=1,inplace=True)
x_train.info()
```

```python
x_train.isnull().sum()
```

```python

x_train["Age"].fillna(x_train["Age"].mean(),inplace=True)

      
      
```

```python
x_train.isnull().sum()
```

```python
type(x_train["Embarked"].mode())
```

```python
x_train.shape
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_g=LabelEncoder()
le_e=LabelEncoder()
```

```python
x_train["Sex"]=le_g.fit_transform(x_train["Sex"])
x_train["Embarked"]=le_e.fit_transform(x_train["Embarked"])
```

```python
x_train.head()
```

```python
x=x_train.iloc[:,1:].values
y=x_train.iloc[:,0].values
x.shape,y.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x[:5]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression(solver='sag',n_jobs=-1)
```

```python
clf.fit(x,y)
```

```python
y_pred=clf.predict(x)
```

```python
clf.score(x,y)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_pred,y))
```

SVM

```python
from sklearn.svm import SVC
```

```python
clf1=SVC(C=100,gamma=1,decision_function_shape='ovo')
```

```python
pred_y1=clf1.predict(x)
```
